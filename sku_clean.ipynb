```python
na_variations = ['n/a', 'na', 'not applicable', 'none', 'n.a', 'n.a.', 'n.a..', 'naa', 'naaa', 'n/a?', 'n/a!', 'N/A']
skuinfo['SKU'] = skuinfo['SKU'].str.lower()

# Replace all 'na' variations with 'na' (case-insensitive)
skuinfo['SKU'].replace(na_variations, 'na', regex=True, inplace=True)

# Convert 'na' back to uppercase in the original column
skuinfo['SKU'] = skuinfo['SKU'].str.upper()
skuinfo['SKU'] = skuinfo['SKU'].apply(lambda x: None if x.lower() == 'na' else x)
skuinfo['SKU'] = skuinfo['SKU'].apply(lambda x: None if x == '' else x)
```

```python
skuinfo['DEPT'] = skuinfo['DEPT'].str.lower()

# Replace all 'na' variations with 'na' (case-insensitive)
skuinfo['DEPT'].replace(na_variations, 'na', regex=True, inplace=True)

# Convert 'na' back to uppercase in the original column
skuinfo['DEPT'] = skuinfo['DEPT'].str.upper()
skuinfo['DEPT'] = skuinfo['DEPT'].apply(lambda x: None if x.lower() == 'na' else x)
skuinfo['DEPT'] = skuinfo['DEPT'].apply(lambda x: None if x == '' else x)
```

```python
skuinfo['CLASSID'] = skuinfo['CLASSID'].str.lower()

# Replace all 'na' variations with 'na' (case-insensitive)
skuinfo['CLASSID'].replace(na_variations, 'na', regex=True, inplace=True)

# Convert 'na' back to uppercase in the original column
skuinfo['CLASSID'] = skuinfo['CLASSID'].str.upper()
skuinfo['CLASSID'] = skuinfo['CLASSID'].apply(lambda x: None if x.lower() == 'na' else x)
skuinfo['CLASSID'] = skuinfo['CLASSID'].apply(lambda x: None if x == '' else x)
```

```python
skuinfo['UPC'] = skuinfo['UPC'].str.lower()

# Replace all 'na' variations with 'na' (case-insensitive)
skuinfo['UPC'].replace(na_variations, 'na', regex=True, inplace=True)

# Convert 'na' back to uppercase in the original column
skuinfo['UPC'] = skuinfo['UPC'].str.upper()
skuinfo['UPC'] = skuinfo['UPC'].apply(lambda x: None if x.lower() == 'na' else x)
skuinfo['UPC'] = skuinfo['UPC'].apply(lambda x: None if x == '' else x)
```

```python
skuinfo['PACKSIZE']=skuinfo['PACKSIZE'].astype(str)
skuinfo['PACKSIZE'] = skuinfo['PACKSIZE'].str.replace('ONE', '1').str.strip()

skuinfo['PACKSIZE'] = skuinfo['PACKSIZE'].str.lower()

# Replace all 'na' variations with 'na' (case-insensitive)
skuinfo['PACKSIZE'].replace(na_variations, 'na', regex=True, inplace=True)

# Convert 'na' back to uppercase in the original column
skuinfo['PACKSIZE'] = skuinfo['PACKSIZE'].str.upper()
skuinfo['PACKSIZE'] = skuinfo['PACKSIZE'].apply(lambda x: None if x.lower() == 'na' else x)
skuinfo['PACKSIZE'] = skuinfo['PACKSIZE'].apply(lambda x: None if x == '' else x)
```
