```python
columnnames=['STORE', 'CITY', 'STATE', 'ZIP_CODE', 'STR_UNKNOWN']
strinfo=pd.read_csv('strinfo.csv',header=None)
strinfo.columns=columnnames
strinfo.drop(['STR_UNKNOWN'],axis=1,inplace=True)
```

```python
import pandas as pd
from uszipcode import SearchEngine

def get_county(zip_code):
    search = SearchEngine()
    result = search.by_zipcode(zip_code)
    if result:
        return result.county
    else:
        return "County not found for this zip code."
# Function to apply to the DataFrame to get county for each zip code
def get_county_for_zip(row):
    return get_county(row['ZIP_CODE'])
# Create a new column 'county' based on 'zip' column
strinfo['COUNTY'] = strinfo.apply(get_county_for_zip, axis=1)
strinfo.drop(['ZIP_CODE'],axis=1,inplace=True)
strfilename='strinfo_sub.csv'
strinfo.to_csv(strfilename)
```
