```python
import pandas as pd
```

```python
skstinfo = pd.read_csv('skstinfo_sub.csv',index_col=0)
strinfo = pd.read_csv('strinfo_sub.csv',index_col=0)
skuinfo = pd.read_csv('skuinfo_sub.csv',index_col=0)
trnsact=pd.read_csv('trnsact_sub.csv',index_col=0)
```

```python
mergeddata=pd.merge(trnsact, skuinfo,on='SKU', how='inner')
```

```python
mergeddata
```

```python
mergeddata_w_skr=pd.merge(mergeddata, skstinfo,on=['SKU','STORE'], how='inner') 
```

```python
mergeddata_w_str=pd.merge(mergeddata_w_skr, strinfo,on='STORE', how='inner') 
```

```python
mergeddata_w_str
```

```python
finalmergename='finalmerge.csv'
```

```python
mergeddata_w_str.to_csv(finalmergename)
```
